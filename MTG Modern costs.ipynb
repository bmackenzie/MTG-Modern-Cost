# Methodology
Data was collected from MTGGoldfish's Modern Meta page (https://www.mtggoldfish.com/metagame/modern/full#paper)

The wayback machine was used to get a snapshot of this page from each year, back to the first available year, 2015. The date selected for each year was the closest available snapshot to 12/21. Notably, this page was not snapshot for the entirety of 2017, so instead the front page of MTGGoldfish's modern meta page was used.  This includes only the top 17 decks, which went down to 2% meta presence for the least popular deck.  It is likely 5-8 decks were missed out in 2017.

I collected the listed paper price for each deck that had a meta presence of at least 1%. This cutoff is arbirtrary, but was used to ensure the data is representative of what the meta looked like in those years, rather than including fringe decks that may influence the data despite seeing very limited play.

I calculated the mean median, max, minimum and count for each year, applied inflation to the deck costs, and calculated those summary statistics again to create the below visualizations.

# Limitations and Assumptions
As noted, the data for 2017 is incomplete as there was no wayback snapshot of the complete modern meta that year.  despite this, data on 17 of the 2017 decks was available, which is likely the majority of the decks that year.

It is assumed that MTGGoldfish's deck and pricing data is accurate.

It is assumed that each snapshot is roughly representative of the entire year. Each snapshot only covers a 14 day period since that was the behavior of MTGGoldfish prior to 2018. It is possible that a meta could become more or less expensive than the snapshotted period within a single year. 

# Conclusions
Modern prices were fiarly consistent from 2015-2019 until a major dip in 2020. The easiest explanation for this dip is that COVID restrictions limited availability of paper play, driving paper prices down until opportunities to play became more available in 2021, where we see a large spike in median deck cost.  The 2021 spike could also be explained by an influx of new cards in Modern Horizons 2, which would be highly desirable both to new and old players.

Minimum deck price is more inconsistent, but 2021 actually appears to be one of the more affordable times to enter modern, assuming you are ok with playing the cheapest decks.  

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cost-stats.csv', header=0)
df
```

```python
inflateddf = pd.read_csv('cost-stats-inflation.csv', header=0)
inflateddf
```

```python
fig, ax = plt.subplots()

ax.plot(df['Year'], df['Median'], marker='o', color='blue')
ax.plot(df['Year'], inflateddf['Median'], marker='o', color = 'red')      
ax.set_xlabel('Year')
ax.set_ylabel('Median Deck Price')
ax.set_title('Median Price of Meta Modern Decks by Year')
plt.legend(('Without Inflation', 'With Inflation'))

plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(df['Year'], df['High'], marker='o', color='blue')
ax.plot(df['Year'], inflateddf['High'], marker='o', color = 'red')      
ax.set_xlabel('Year')
ax.set_ylabel('Highest Deck Price')
ax.set_title('Highest Price Meta Modern Deck by Year')
plt.legend(('Without Inflation', 'With Inflation'))

plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(df['Year'], df['Low'], marker='o', color='blue')
ax.plot(df['Year'], inflateddf['Low'], marker='o', color = 'red')      
ax.set_xlabel('Year')
ax.set_ylabel('Lowest Deck Price')
ax.set_title('Lowest Price Meta Modern Deck by Year')
plt.legend(('Without Inflation', 'With Inflation'))

plt.show()
```
